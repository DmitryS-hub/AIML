#### Define Utility Functions

```python
def plot_counts_percentages(data, col, vr=0):
    total_clients = data.shape[0]
    chart = sns.countplot(data[col], order = data[col].value_counts().index, palette='muted');
    chart.set(title=col.title(), xlabel='', ylabel='Count')
    chart.set_xticklabels(chart.get_xticklabels(),  rotation=vr)
    for p in chart.patches:
        height = p.get_height()
        chart.text(p.get_x() + p.get_width()/2., height + 3, '{:1.2f}'.format((height/total_clients)*100), ha='center')

def display_num_stat_graph(data, col, hist_color='seagreen', hist_mean='tomato', hist_median='steelblue', box_color='steelblue'):
    f = plt.figure(figsize=(15,5))
    gs = f.add_gridspec(1, 2)

    with sns.axes_style("whitegrid"):
        ax = f.add_subplot(gs[0, 0])
        sns.distplot(data[col], color=hist_color, kde=True, rug=True)
        ax.axvline(data[col].mean(), color = hist_mean, label='Mean')
        ax.axvline(data[col].median(), color = hist_median, label='Median')
        ax.legend()

    with sns.axes_style("darkgrid"):
        ax = f.add_subplot(gs[0, 1])
        sns.boxplot(x=data[col], notch = True, whis = 1.5, color=box_color)
    
    f.tight_layout()
    plt.show()
    
def printc(text, color):
    colors = {
        'black':'\033[30m',
        'red':'\033[31m',
        'green':'\033[32m',
        'orange':'\033[33m',
        'blue':'\033[34m',
        'purple':'\033[35m',
        'cyan':'\033[36m',
        'lightgrey':'\033[37m',
        'darkgrey':'\033[90m',
        'lightred':'\033[91m',
        'lightgreen':'\033[92m',
        'yellow':'\033[93m',
        'lightblue':'\033[94m',
        'pink':'\033[95m',
        'lightcyan':'\033[96m'
    }
    style = {
        'reset':'\033[0m',
        'bold':'\033[01m'
    }
    
    print(colors[color], text, style['reset'])
        

def display_num_skew_outliers(data, col):
    # Computing IQR
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    LowerWhisker = Q1 - 1.5 * IQR
    MinValue = data[col].min()
    UpperWhisker = Q3 + 1.5 * IQR
    MaxValue = data[col].max()
    
    # Update whiskers
    LowerWhisker = max(MinValue, LowerWhisker)
    UpperWhisker = min(MaxValue, UpperWhisker)
    
    # Outliers
    LowerOutliers = data[data[col] < LowerWhisker][col]   
    LowerOutliersCount = LowerOutliers.count()
    UpperOutliers = data[data[col] > UpperWhisker][col]
    UpperOutliersCount = UpperOutliers.count()

    print('>', col.title())
    printc('Mean         : {}'.format(data[col].mean()), 'red') 
    printc('Median       : {}'.format(data[col].median()), 'blue') 
    print('Skew          :', data[col].skew())
    print('Lower whisker :', LowerWhisker)
    print('Upper whisker :', UpperWhisker)

    print('Outliers above upper whisker: ', UpperOutliersCount, '({:.1f}%)'.format( (UpperOutliersCount/data.shape[0])*100) )
    if(UpperOutliersCount > 0):
        print('  > (closest 15):', ', '.join(map(str, UpperOutliers.head(15))))
    
    print('Outliers under lower whisker: ', LowerOutliersCount, '({:.1f}%)'.format( (LowerOutliersCount/data.shape[0])*100) )
    if(LowerOutliersCount > 0):
        print('  > (closest 15):', ', '.join(map(str, LowerOutliers.head(15))))
```

## <a id = "ann"></a> 3. Building ANN model

So now let us build our ANN model. We know that the bank wants to find out whether clients will leave or not. Also, we know that the bank wants to retain those clients by identifying problems and concentrating on the improvement of its services. In the classification aspect, we want to build a model that performs well in predicting people who may leave and at the same time has minimum **Type II error** i.e. we want to minimize **False Negatives**. False positives are less of an issue because they will be looked at by the bank employees where they can further reduce the error, in contrast, false negatives will be never addressed as they are considered safe by the model. Hence our main metric that we should look at is **Recall**.

Furthermore, we know that our **dataset is imbalanced** towards the clients who stay, this will lead to a model that very good at predicting clients who won't leave resulting in **high values for true negatives**. This means that we cannot rely on **Accuracy** metric as it will always be influenced by the majority class and be very high.

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
classifier = Sequential()
#initializer = tf.keras.initializers.Ones()
#initializer = tf.keras.initializers.LecunUniform() 
#initializer = tf.keras.initializers.LecunNormal()
#initializer = tf.keras.initializers.HeNormal()
#initializer = tf.keras.initializers.HeUniform()
initializer = tf.keras.initializers.GlorotUniform() #'uniform'

# Input layer and first hidden layer
classifier.add(Dense(units=12, kernel_initializer=initializer, activation='relu', input_dim=X_train.shape[1]))

# Second hidden layer
classifier.add(Dense(units=6, kernel_initializer=initializer, activation='relu')) #tanh

# Output layer
classifier.add(Dense(units=1, kernel_initializer=initializer, activation='sigmoid'))
```

```python
# Compiling the ANN model
#optimizer = tf.keras.optimizers.RMSprop(learning_rate=0.01)
#optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
classifier.compile(loss='binary_crossentropy', optimizer=optimizer , metrics=['accuracy', 'Recall', 'Precision', 'AUC'])

# Fitting the ANN to the Training set
history = classifier.fit(X_train, y_train, batch_size=20, epochs=100, validation_split=0.2, verbose=0)
```

```python
hist  = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

### <a id = "ann_review"></a>Model review

```python
column = True
for colset in [['accuracy', 'val_accuracy'], ['recall', 'val_recall'], ['precision', 'val_precision'], ['auc', 'val_auc'], ['loss', 'val_loss']]:

    if(column):
        plt.figure(figsize=(15,5))

    plt.subplot(1, 2, 1 if column else 2)
    plt.plot(hist[colset[0]])
    plt.plot(hist[colset[1]])
    plt.title(colset[0].capitalize())
    plt.legend(("Training" , "Validation") , loc=0)

    if(not column):
        plt.show()
    column = not column
```

```python
classifier.summary()
```

```python
classifier_evaluation = pd.DataFrame(
        [classifier.evaluate(X_train, y_train, verbose=0), classifier.evaluate(X_test, y_test, verbose=0)],
        columns=classifier.metrics_names,
        index=['Training', 'Testing']
    )

display(classifier_evaluation)
```

**Insights:**
 - The output of the model varies every run as our model weights are randomly initialized. This makes it difficult to pinpoint the best hyperparameters when it comes to small improvements.
 - We have tried various ANN configurations - deep & shallow with a various number of neurons on each layer, different activation functions, initialization distributions, learning rates and optimizers. The best that we could get is **Recall ~0.46** and **Loss ~0.34** on the test dataset without significant overfitting. This possibly points out the fact that we are stuck in a local minimum and cannot completely fit the data. Increasing the number of hidden layers and/or neurons in order to improve decision boundary did not help to improve the situation.
 - For this dataset a small ANN of 2 hidden layers (12 and 6 neurons respectively) and simple **Relu** activation function will perform almost the same to a more complex 3-4 hidden layers with 32-8 neurons per layer and **Tanh** activation function.
 - Learning rate however drastically affects the abbility of the model to converge. Learning rate of **0.01** showed the best results.
 - **SGD** optimizers also performed the best keeping model from overfitting the longest.

### <a id = "ann_scores_cm"></a>Scores and Confusion matrix 

#### Define Utility Functions

```python
from sklearn.metrics import roc_auc_score, roc_curve

def report_ROC(estimator, x_te, y_te, roc_auc, title, force_show=True):
    y_pred = estimator.predict(x_te)
    fpr, tpr, thresholds = roc_curve(y_te, y_pred)
    plt.plot(fpr, tpr, label='{} (area = {:0.2f})'.format(title, roc_auc))
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic')
    plt.legend(loc="lower right")
    if(force_show):
        plt.show()
        
    return {'fpr': fpr, 'tpr':tpr, 'thresholds': thresholds, 'roc_auc': roc_auc}
```

```python
from sklearn.metrics import confusion_matrix

def cm_report(cm, title, report, force_show=True):
    flat_cm = cm.flatten();
    group_names = ['TN', 'FP', 'FN', 'TP']
    group_counts = ['{}'.format(value) for value in flat_cm]
    labels = [f'{v1}\n{v2}' for v1, v2 in zip(group_names, group_counts)]
    labels = np.asarray(labels).reshape(2,2)
    color_map = { flat_cm[0]:'mediumseagreen', flat_cm[1]:'tomato', flat_cm[2]:'tomato', flat_cm[3]:'mediumseagreen'}
    colors =  [(color_map[index] if index in color_map.keys() else 'black') for index in np.arange(flat_cm.min(), flat_cm.max() + 1)] 
    
    #plt.figure(figsize = (7,5))
    sns.heatmap(cm, annot=labels, fmt='', xticklabels = [0, 1], yticklabels = [0, 1], cmap=colors, cbar = False, annot_kws={"size": 20})
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    if(force_show):
        plt.show()
    
    TP = cm.flatten()[3]
    TN = cm.flatten()[0]
    FP = cm.flatten()[1]
    FN = cm.flatten()[2]

    recall = TP/(TP + FN)
    specificity = TN/(FP + TN)
    precision = TP/(TP + FP)
    accuracy = (TP + TN)/(TP + TN + FP + FN)
    error_rate = 1 - accuracy
    
    report['Accuracy'] = accuracy
    report['Precision'] = precision
    report['Recall/Sensitivity/TPR'] = recall
    report['Specificity/TNR'] = specificity
    report['Error Rate'] = error_rate
    report['TP'] = TP
    report['TN'] = TN
    report['FP'] = FP
    report['FN'] = FN
    
    return report
```

```python
from sklearn.metrics import recall_score, precision_score, f1_score, accuracy_score, classification_report

def estimator_report(estimator, title, x_te, y_te):
    y_pred = (estimator.predict(x_te) > 0.5).astype("int32")
    f1 = f1_score(y_te, y_pred)
    cm = confusion_matrix(y_te, y_pred)
    roc_auc = roc_auc_score(y_te, y_pred)
    
    estimator_cm_report = pd.DataFrame(
        [[-1, -1, -1, -1, roc_auc, f1, -1, -1, -1, -1, -1]],
        columns=['Accuracy', 'Precision', 'Recall/Sensitivity/TPR', 'Specificity/TNR', 'ROC AUC', 'F1', 'Error Rate', 'TP', 'TN', 'FP', 'FN'],
        index=[title]
    )

    # Confusion matrix & ROC AUC
    plt.figure(figsize=(15,5))

    plt.subplot(1, 2, 1)
    estimator_cm_report = cm_report(cm, title, estimator_cm_report, False)

    plt.subplot(1, 2, 2)
    roc_report = report_ROC(estimator, x_te, y_te, roc_auc, title, force_show=False)
   
    plt.show()
    
    return {'cm': estimator_cm_report, 'roc': roc_report, 'title': title}
```
